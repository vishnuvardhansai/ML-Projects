```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data = pd.read_csv('/content/sonar_data.csv',header=None)
```

```python
sonar_data.head()
```

```python
sonar_data.tail()
```

```python
sonar_data.describe()
```

```python
x = sonar_data.drop(columns=60,axis=1)
y = sonar_data[60]
```

```python
print(x)
```

```python
print(y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
print(x.shape,X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(Y_train)
```

```python
print(Y_test)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#Finding The Training data Accuracy
```

```python
X_train_prediction = model.predict(X_train)
X_train_prediction_accuracy = accuracy_score(X_train_prediction,Y_train)
print('Accuracy of Training Data : ',X_train_prediction_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
X_test_prediction_accuracy = accuracy_score(X_test_prediction,Y_test)
print('Accuracy of Test Data : ',X_test_prediction_accuracy)
```
